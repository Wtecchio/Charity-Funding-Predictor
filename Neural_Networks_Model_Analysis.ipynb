## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
appDF = pd.read_csv("Resources/charity_data.csv")
appDF.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
appDF = appDF.drop(columns=["EIN", "NAME"], axis=1)
appDF.head()
```

```python
# Determine the number of unique values in each column.
appDF.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
appDF["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

#cut off will be the T8 application type (737)
application_types_to_replace = appDF[appDF['APPLICATION_TYPE'].map(appDF['APPLICATION_TYPE'].value_counts()) < 737]
application_types_to_replace = application_types_to_replace['APPLICATION_TYPE'].unique().tolist()

# Replace in dataframe
for app in application_types_to_replace:
    appDF['APPLICATION_TYPE'] = appDF['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
appDF['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
appDF['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
bucket = appDF[appDF['CLASSIFICATION'].map(appDF['CLASSIFICATION'].value_counts()) > 1]
bucket['CLASSIFICATION'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced (C5000 - 116)
# use the variable name `classifications_to_replace`
classifications_to_replace = appDF[appDF['CLASSIFICATION'].map(appDF['CLASSIFICATION'].value_counts()) < 116]
classifications_to_replace = classifications_to_replace['CLASSIFICATION'].unique().tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    appDF['CLASSIFICATION'] = appDF['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
appDF['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies = pd.get_dummies(appDF)
dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
temp_X = dummies.drop('IS_SUCCESSFUL', axis=1)
temp_Y = dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(temp_X, temp_Y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

neuralNetwork = tf.keras.models.Sequential()

# First hidden layer
neuralNetwork.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=45)) #input_dim means how many columns are in the data frame

# Second hidden layer
neuralNetwork.add(tf.keras.layers.Dense(units = 3, activation="tanh"))

# Output layer
neuralNetwork.add(tf.keras.layers.Dense(units=1, activation = "sigmoid"))

# Check the structure of the model
neuralNetwork.summary()
```

```python
# Compile the model
neuralNetwork.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = fit_model = neuralNetwork.fit(X_train_scaled, Y_train, epochs=50)
```
